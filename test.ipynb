```python
import cv2
from seam import SeamCarve
import matplotlib.pyplot as plt
```

```python

img = cv2.imread('./image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print('Image :')
plt.imshow(img)
```

SEAM CARVING 1: RESIZE IMAGE

```python
sc_img = SeamCarve(img)
print('shape image before: ',  sc_img.image().shape)
sc_img.resize(new_height=300, new_width=300)   # resize image
plt.imshow(sc_img.image())
```

SEAM CARVING 2: REMOVE A MASK

```python
mask = cv2.imread('./mask.jpg', 0) != 255
sc2_img = SeamCarve(img)
sc2_img.remove_mask(mask)

plt.imshow(sc2_img.image())
```

SEAM CARVING 3: REMOVE A MASK BUT NOT RESIZE

```python
sc3_img = SeamCarve(img)
old_width, old_height = img.shape[0], img.shape[1]
sc3_img.remove_mask(mask)
sc3_img.resize(new_height=old_height, new_width=old_width)
plt.imshow(sc3_img.image())
```
